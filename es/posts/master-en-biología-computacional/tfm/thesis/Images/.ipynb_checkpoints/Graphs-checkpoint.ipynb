```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv("./test_runs.csv")
# make fig
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

# Change in average temperature before/after industrial revolution
# df[df.apply(lambda x: np.abs(x - x.mean()) / x.std() < 0.5).all(axis=1)]

#fig.suptitle('Procedence of the data in our outputs:\n by Data Source and by Import Method', size=25)

sns.set_theme(style="ticks", palette="pastel")
ax1 = sns.boxplot(data=df[["InChI",
                          "ChEBI","Structural Similarity","Name",
                          "HMDB_ID"]], ax=ax1, palette=["m", "y", "g", "b", "r"], dodge=False)
ax1 = sns.stripplot(data=df[["InChI",
                          "ChEBI","Structural Similarity","Name",
                          "HMDB_ID"]], color="#828282", ax=ax1)
ax1.set_ylim(0,45)
ax1.set_xlabel("Categories used for data Import")
ax1.set_ylabel("Number of times used per run")

ax2 = sns.boxplot(data=df[["HMDB",
                          "WikiData","SMPDB","Name",
                          "HMDB_ID"]], ax=ax2, palette=["#FFD9C0", "#F9F9F9", "#FAF4B7", "pink", "#CDF0EA"], dodge=False)
ax2 = sns.stripplot(data=df[["HMDB",
                          "WikiData","SMPDB","Name",
                          "HMDB_ID"]], color="#828282", ax=ax2)
ax2.set_ylim(0,150)
ax2.set_xlabel("Databases from which data has been Impòrted")
ax2.set_ylabel("Number of times used per run")

plt.savefig("./figure_4.png", dpi=800)
```
